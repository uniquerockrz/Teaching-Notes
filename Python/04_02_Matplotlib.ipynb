```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line Plot

```python
plt.plot([1, 2, 2, 5, 7, 8 , 9, 1])
```

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
```

```python
X = np.random.randint(0, 100, 100)
```

```python
X
```

```python
Y = np.random.randint(0, 100, 100)
```

```python
Y
```

```python
plt.plot(X, color='red')
```

```python
plt.plot(Y, color='green')
```

```python
plt.plot(X, color='red')
plt.plot(Y, color='green')
```

```python
plt.plot([1, 5, 3, 2, 5], [10, 11, 12, 13, 14], marker='o', linewidth=3, label='X')
plt.plot([1, 7, 2, 2, 1], [10, 11, 12, 13, 14], marker='o', linewidth=3, label='Y')
plt.xlabel('X Asis')
plt.ylabel('Y Asis')
plt.title('X over Y graph', fontsize=22)
plt.legend()
```

### Pie Chart

```python
plt.pie([1, 2 ,3, 4])
```

### Histograms

```python
plt.hist(X, bins=5, color='red', label='X')
plt.hist(Y, bins=5, color='blue', label='Y')
plt.legend()
```

### Scatter Plots

```python
plt.scatter(X, Y, alpha=0.5)
```

### Bar Charts

```python
plt.bar(X, Y)
```
