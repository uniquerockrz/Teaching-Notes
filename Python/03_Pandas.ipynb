### Group By

```python
df_students = pd.DataFrame({
    'names': ['Jack', 'John', 'Simon', 'Rachel', 'Angela', 'Michel', 'Sarah', 'Suzie', 'Dan', 'Andrew'],
    'grades': ['A', 'B', 'B', 'C', 'A', 'D', 'C', 'B', 'A', 'C']
})
```

```python
df_students.head(10)
```

```python
df_students.groupby('grades').groups
```

```python
df_students.groupby('grades').get_group('A')
```

```python
df_students.groupby('grades').get_group('B')
```

```python
df_students.groupby('grades').count()
```

```python
df_students['grades'].value_counts()
```

### Applying Functions

```python
df_countries.head(10)
```

```python
df_countries['Population'] / 1000
```

```python
def divide_by_million(data, column):
    data_to_divide = data[column]
    return round(data_to_divide / 1000000, 2)
```

```python
df_countries['Population in millions'] = df_countries.apply(divide_by_million, args=('Population',), axis=1)
```
