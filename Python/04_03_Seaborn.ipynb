```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.randint(0, 100, 20)
Y = np.random.randint(0, 100, 20)
```

```python
X
```

```python
Y
```

### Scatterplot

```python
sns.scatterplot(x=X, y=Y, marker='+')
plt.title('Scatterplot')
```

```python
sns.regplot(x=X, y=Y)
plt.title('Regression fit plot')
```

### Line Plot

```python
sns.lineplot(x=X, y=Y)
```

### Histogram Plot

```python
sns.displot(X)
```

```python
sns.displot(X, bins=10)
```

### Density Plot

```python
sns.kdeplot(X)
```

```python
sns.kdeplot(X, fill=True)
```

```python
sns.kdeplot(X, fill=True, color='red')
sns.kdeplot(Y, fill=True, color='green')
```

### Bar Chart

```python
sns.barplot(x=X, y=Y)
```

```python
sns.barplot(x=X, y=Y, orient='h')
```

### Box Plot

```python
sns.boxplot(X)
```

### Violin Plot

```python
sns.violinplot(X)
```
